# Tkinter-屏保
### 项目分析
- 屏保可以自己启动，也可以手动启动
- 一旦敲击键盘或者移动鼠标，或者其他引发时间，则停止
- 如果屏保是一幅画的话，则没有画框
- 图像的动作是随机的,具有随机性，可能包括颜色，大小，方向
- 世界构成
    - ScreenSaver：
        - 需要一个canvas， 大小与屏幕一致，没有边框
    - Ball
        - 颜色，大小，方向
        - 球能东，可以被调用
