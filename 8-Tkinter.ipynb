### 组件布局
- 控制组件的摆放方式
- 三种布局：
    - pack： 按照方位布局
    - place： 按照坐标布局
    - grid： 网格布局
    
- pack布局
    - 最简单，代码最少，按个排放，默认从上倒下，系统自动设置
    - 使用方式： 组件对象.pack（设置，，，）
    - side： 停靠的方位，可以选值为LEFT,TOP,RIGHT,BOTTON
    - fill: 填充方式，X，Y,BOTH,NONE
    - EXPANDE：YES/NO
    - anchor: N,E,S,W,CENTER
    - ipadx:x方向的内编剧
    - ipady:y方向的内布局
    - padx：x方向外边界
    - pady：y方向外边界
- grid布局
    - 通用使用方式：组件对象.grid（设置，，，）
    - 利用row，column编号，都是从0开始
    - sticky： N，E，S，W表示上下左右，用来绝对组件从哪个方向开始
    - 支持ipadxpadx等参数，与pack函数含义意义
    - 支持rowspan， columespan，表示跨行，跨咧数量
- place布局：
    - 明确方位的摆放
    - 相对位置布局，随意改变窗口大小会导致混乱
    - 使用palce函数，分为绝对布局和相对布局，绝对布局使用x，y参数
    - 相对布局使用relx，rely，relheight，relwidth

```python
# pack布局案例
import tkinter

baseFrame = tkinter.Tk()

btn1 = tkinter.Button(baseFrame, text="1111111")
btn1.pack(side=tkinter.LEFT, expand=tkinter.YES, fill=tkinter.BOTH)

btn1 = tkinter.Button(baseFrame, text="2222222")
btn1.pack(side=tkinter.TOP, expand=tkinter.YES, fill=tkinter.BOTH)

btn1 = tkinter.Button(baseFrame, text="333333333")
btn1.pack(side=tkinter.RIGHT, expand=tkinter.YES, fill=tkinter.NONE, anchor=tkinter.NE)

btn1 = tkinter.Button(baseFrame, text="44444444")
btn1.pack(side=tkinter.LEFT, expand=tkinter.NO, fill=tkinter.Y)

btn1 = tkinter.Button(baseFrame, text="55555555")
btn1.pack(side=tkinter.TOP, expand=tkinter.NO, fill=tkinter.BOTH)

btn1 = tkinter.Button(baseFrame, text="6666666")
btn1.pack(side=tkinter.BOTTOM, expand=tkinter.YES)

btn1 = tkinter.Button(baseFrame, text="7777777")
btn1.pack(anchor=tkinter.SE)



baseFrame.mainloop()
```

## 消息机制
- 消息的传递机制
    - 自动发出事件/消息
    - 消息由系统负责发送到队列
    - 由相关组件进行绑定/设置
    - 后端自动选择感兴趣的事件并做出相应反应
- 消息格式：
    - <[modifier-]---type-[-detail]>
    - <Button-1>: Button表示一个按钮事件，1代表的是鼠标左键
    - <KeyPress-A>: 键盘A键位
    - <Control-shift-KeyPress-A>:同时按下ctrl+shift+A
    - <F1>:F1键盘
    - [键位对应名称]

```python
# 事件的简单例子
import tkinter

def baseLabel(evennt):
    global baseFrame
    print("点击了")
    lb = tkinter.Label(baseFrame, text="谢谢点击")
    lb.pack()
    
#　画出程序总框架
baseFrame = tkinter.Tk()

lb = tkinter.Label(baseFrame, text="模拟按钮")
# label绑定相应的消息和处理函数
# 自动获取左键点击
lb.bind("<Button-3>", baseLabel)
lb.pack()

baseFrame.mainloop()
```

### Tkinter的绑定
- bind_all:全局范围的绑定，默认的是全局快捷键，比如F1是帮助文档
- bind_class: 接受三个参数，第一个是类名，第二个是事件，第三个是操作
    - w.bind_class("Entry", '<Control-V>', my_paste)
- bind:单独对某一个实例绑定
- unbind：解绑，需要一个参数，即你要解绑哪个事件

### 菜单
### 1. 普通菜单
- 第一个Menu类定义的是parent
- add_command 添加菜单项，如果惨淡是顶层菜单，则从左向右添加，否则就是下拉菜单
    - label： 指定菜单项名称
    - command： 点击后相应的调用函数
    - acceletro: 快捷键
    - underline: 制定是否菜单信息下有横线
    - menu： 属性指定使用哪一个作为顶层菜单

```python
# 普通菜单的代码
import tkinter

baseFrame = tkinter.Tk()

menubar = tkinter.Menu(baseFrame)

for item in ["File", 'Edit', 'View']:
    menubar.add_command(label=item)
    
baseFrame['menu'] = menubar


baseFrame.mainloop()
```

### 级联菜单
- add_cascade:级联菜单，作用是引出后面的菜单
- add_cascade的menu属性：指明把菜单级联到哪个菜单上
- label：名称
- 过程：
    - 1.建立menu实例
    - 2.add_command
    - 3.add_cascade

```python
import tkinter

baseFrame = tkinter.Tk()

menubar = tkinter.Menu(baseFrame)

emenu = tkinter.Menu(menubar)
for item in ["copy", 'past', 'cut']:
    emenu.add_command(label=item)
    
menubar.add_cascade(label="File")
menubar.add_cascade(label='Edit', menu = emenu)
menubar.add_cascade(label='View')

baseFrame['menu'] = menubar

baseFrame.mainloop()
```

### 弹出式菜单
- 弹出菜单也叫上下文菜单
- 实现的大致思路
    1. 建立菜单并向菜单添加各种功能
    2. 监听鼠标右键
    2. 如果右键点击，则根据位置判断弹出
    4. 调用Menu的pop方法
- add_separator:添加分隔符

```python
import tkinter 

def makeLabel():
    global baseFrame
    tkinter.Label(baseFrame, text="PHP是最好的编程语音，我用Python").pack()

baseFrame = tkinter.Tk()

menubar = tkinter.Menu(baseFrame)
for x in ["copy", 'past', 'c俺是个']:
    menubar.add_separator()
    menubar.add_command(label=x)
menubar.add_command(label='重庆火锅', command=makeLabel)

# 事件处理函数一定要至少有一个参数，且第一个参数表示的是系统事件
def pop(event):
    # 
    menubar.post(event.x_root, event.y_root)

baseFrame.bind("<Button-3>", pop)


baseFrame.mainloop()
```

### canvas 画布
- 画布： 可以自由在上面绘制图像的小舞台
- 在画布上绘制对象，通常用create_xxxx, xxxx=对象类型， 例如line，rectangle
- 画布的作用是把一定组件画到画布上显示出来
- 画布支持的组件
    - arc
    - bitmap
    - image
    - line
    -oval
    - polygon
    - rectangle
    - text
    - window
- 每次调用creatx_xxx都会返回一个创建组件的ID，同事可以使用tag属性指定其标签
- 通过调用canvas.move实现一个简单动画

```python
import tkinter 

baseFrame = tkinter.Tk()

cvs = tkinter.Canvas(baseFrame, width=300, height=200)
cvs.pack()
# 一条线需要两个点
cvs.create_line(22,23,190,234)
cvs.create_text(56,67, text="耶耶耶耶耶耶耶耶耶")


baseFrame.mainloop()
```
